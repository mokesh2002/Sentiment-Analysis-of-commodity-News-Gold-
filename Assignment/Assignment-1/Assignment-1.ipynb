```python
def countplot(dataframe, x_val, plot_title='', figsize=(9,4)):
    plt.figure(figsize=figsize)
    plt.title(plot_title)
    sns.countplot(data=df, x=x_val)
    plt.show()
```

```python
def barplot(dataframe, x_val, y_val):
    sns.barplot(data=dataframe, x=x_val, y=y_val)
    plt.title(x_val.title() + ' vs ' + y_val.title())
    plt.show()
```

Multi-Variate Analysis

```python
fig = plt.figure(figsize = [30, 18])
for i in range(1, len(list(df.columns))):
    plt.subplot(8, 3, i) 
    
    fig.tight_layout(pad=2)
    df_new = df.iloc[:,i] 
    
    plt.suptitle('Comparision of all Features') 
    plt.title(list(df.columns)[i]) 
    
    sns.boxenplot(df_new) 

plt.show()
```
